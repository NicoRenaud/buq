```python
import numpy as np
from buq.buq import ClassicalBayesianQuadratureMD
from buq.data.md import AlanineDipeptideMD
from buq.kernels.rbf import RBFKernel
```

```python
target_function = AlanineDipeptideMD(mode='debug')
```

```python
rbf = RBFKernel(lengthscale=0.75, noise=0.0)
```

```python
x_data = target_function.generate_samples(10)
y_data = target_function(x_data)
```

```python
bq = ClassicalBayesianQuadratureMD(rbf, target_function)
```

```python
res = bq.run(10, plot_fit=False)
```

```python
import matplotlib.pyplot as plt
plt.plot(res.rmse)
```
